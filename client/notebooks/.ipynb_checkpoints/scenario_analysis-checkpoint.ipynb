```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
GRID_SIZE = 64
GRID_SIZE_2 = GRID_SIZE ** 2
TRIALS = 10
TURNS = 81
DRONES = 32
PHI = 1.0
VICTIMS = 1024
```

```python
def upperbound(x, grid_size=GRID_SIZE, turns=TURNS, phi=PHI):
    grid_size_2 = GRID_SIZE * GRID_SIZE
    samples, weight_per_sample, total_acc_weight = x * turns, 1, 0
    
    while samples > 0:
        measured_weight = weight_per_sample * phi
        total_acc_weight += min(grid_size_2, samples) * measured_weight
        weight_per_sample -= measured_weight
        samples -= min(grid_size_2, samples)
    
    return total_acc_weight / grid_size_2
```

```python
report_headers = ['Scenario', 'Alpha', 'Beta', 'Gamma', 'Drones', 'Victims', 'Score']
groupby_headers = report_headers[:5]
```

```python
data_randomwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32RandomWalkA.csv')
data_weightedwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32WeightedWalkA.csv')
data_smgwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32SmgWalkA.csv')
data_quadsmg_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32QuadSmgA.csv')
data_globalexplore_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32GlobalExploreA.csv')
data_bestwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32BestWalkA.csv')
```

```python
report_randomwalk_64_32 = data_randomwalk_64_32[report_headers].groupby(groupby_headers).mean().reset_index()
report_weightedwalk_64_32 = data_weightedwalk_64_32[report_headers].groupby(groupby_headers).mean().reset_index()
report_smgwalk_64_32 = data_smgwalk_64_32[report_headers].groupby(groupby_headers).mean().reset_index()
report_quadsmg_64_32 = data_quadsmg_64_32[report_headers].groupby(groupby_headers).mean().reset_index()
report_globalexplore_64_32 = data_globalexplore_64_32[report_headers].groupby(groupby_headers).mean().reset_index()
report_bestwalk_64_32 = data_bestwalk_64_32[report_headers].groupby(groupby_headers).mean().reset_index()
```

```python
x_labels = ['coverage', 'default', 'weighted', 'weighted; min coll.', 'weighted; min coll.; quad disp.', 'global assign', 'max; min coll.; quad disp.']
x_colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown', 'pink']
```

```python
x_axis = np.arange(len(x_labels))
PHI = 1

plt.bar(x_axis[0], np.array(list(map(lambda _: upperbound(DRONES, phi=PHI), x_axis))), label=x_labels[0], color=x_colors[0])
plt.bar(x_axis[1], report_randomwalk_64_32.loc[report_randomwalk_64_32['Beta'] == PHI]['Victims'] / VICTIMS, label=x_labels[1], color=x_colors[1])
plt.bar(x_axis[2], report_weightedwalk_64_32.loc[report_weightedwalk_64_32['Beta'] == PHI]['Victims'] / VICTIMS, label=x_labels[2], color=x_colors[2])
plt.bar(x_axis[3], report_smgwalk_64_32.loc[report_smgwalk_64_32['Beta'] == PHI]['Victims'] / VICTIMS, label=x_labels[3], color=x_colors[3])
plt.bar(x_axis[4], report_quadsmg_64_32.loc[report_quadsmg_64_32['Beta'] == PHI]['Victims'] / VICTIMS, label=x_labels[4], color=x_colors[4])
plt.bar(x_axis[5], report_globalexplore_64_32.loc[report_globalexplore_64_32['Beta'] == PHI]['Victims'] / VICTIMS, label=x_labels[5], color=x_colors[5])
plt.bar(x_axis[6], report_bestwalk_64_32.loc[report_bestwalk_64_32['Beta'] == PHI]['Victims'] / VICTIMS, label=x_labels[6], color=x_colors[6])

plt.title(f'Mean % Victims Located based on Strategy (Grid Size={GRID_SIZE}; Turns={TURNS}, Drones={DRONES}; Victims={VICTIMS}; Alpha=1.00, Phi={PHI:.2f})')
plt.xticks(x_axis, labels=x_labels, rotation=45)
plt.ylabel('Mean % Found')
```
