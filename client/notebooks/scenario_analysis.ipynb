```python
def random_placement(probes=VICTIMS, grid_size=GRID_SIZE, inc=20):
    placements = [(rng.uniform(0, 1) * grid_size, rng.uniform(0, 1) * grid_size) for _ in range(probes)]
    
    grid_center = grid_size / 2
    distances = [math.sqrt((X[0] - grid_center)**2 + (X[1] - grid_center)**2) for X in placements]
    distances.sort()
    
    counts, i, k = [0], 0, 0
    while i < len(distances):
        if distances[i] > k and not math.isclose(distances[i], k):
            counts.append(counts[-1])
            k += inc
            continue
        
        counts[-1] += 1
        i += 1
    
    return (counts, inc)
```

```python
counts = random_placement(grid_size=GRID_SIZE * 14)[0]
counts = list(map(lambda x: x / VICTIMS, counts))
```
