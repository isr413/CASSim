```python
def utility(df, alpha, gamma):
    victims = df.loc[df['Alpha'] == alpha].loc[df['Gamma'] == gamma]['Victims'] / VICTIMS
    score = df.loc[df['Alpha'] == alpha].loc[df['Gamma'] == gamma]['Score'] / VICTIMS
    return victims + score
```

```python
x_axis = np.arange(82)

plt.plot(x_axis, utility(report_selfish_64_32, 0.875, 0.75), label="Selfish", color=x_colors[0])
plt.plot(x_axis, utility(report_generous_64_32, 0.875, 0.75), label="Generous", color=x_colors[7])
plt.plot(x_axis, utility(report_helpful_64_32, 0.875, 0.75), label="Helpful", color=x_colors[2])
plt.plot(x_axis, utility(report_eager_64_32, 0.875, 0.75), label="Eager", color=x_colors[3])
plt.plot(x_axis, report_quad_64_32.loc[report_quad_64_32['Alpha'] == 0.875].loc[report_quad_64_32['Beta'] == 0.75]['Victims'] / VICTIMS, label=x_labels[6], color=x_colors[6])

plt.title(f'Mean % Utility based on Attitude (Alpha=0.875, Phi=0.75)')
plt.xlabel('Turns')
plt.ylabel('Mean % Utility')
plt.yticks(np.arange(0, 0.55, step=0.05))
plt.legend()
```
