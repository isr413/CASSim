```python
import matplotlib as mpl

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from matplotlib import cm
from matplotlib.ticker import MaxNLocator
```

```python
VICTIMS = 1024
```

```python
data_randomwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32RandomWalkB.csv')
data_weightedwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32WeightedWalkB.csv')
data_smgwalk_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32StochasticWalkB.csv')
data_allowcolls_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32SmgWalkB.csv')
data_nocolls_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32BestWalkB.csv')
data_global_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32GlobalExploreB.csv')
data_quad_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32QuadSmgB.csv')
data_cd6_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32QuadSmgC.csv')
data_cd3_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32QuadSmgD.csv')
data_cd9_64_32 = pd.read_csv('../logs/parser_ScenarioSize64Probes32QuadSmgE.csv')
```

```python
data = data_randomwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()
fig = plt.figure(figsize=(13.33,7.5), dpi = 96)
ax = plt.axes(projection='3d')
ax.set_box_aspect(aspect=None, zoom=0.9)
ax.set_title('Victims Rescued by Alpha and Beta for Random Walk')
ax.set_xlabel('Alpha')
ax.set_ylabel('Beta')
ax.set_zlabel('Victims Rescued')
surf = ax.plot_trisurf(data['Alpha'], data['Beta'], data['Victims'] / VICTIMS, cmap=cm.coolwarm, edgecolor='none', linewidth=0, antialiased=False)
```

```python
data = data_weightedwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()
fig = plt.figure(figsize=(13.33,7.5), dpi = 96)
ax = plt.axes(projection='3d')
ax.set_box_aspect(aspect=None, zoom=0.9)
ax.set_title('Victims Rescued by Alpha and Beta for Greedy Walk')
ax.set_xlabel('Alpha')
ax.set_ylabel('Beta')
ax.set_zlabel('Victims Rescued')
surf = ax.plot_trisurf(data['Alpha'], data['Beta'], data['Victims'] / VICTIMS, cmap=cm.coolwarm, edgecolor='none', linewidth=0, antialiased=False)
```

```python
data = data_smgwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()
fig = plt.figure(figsize=(13.33,7.5), dpi = 96)
ax = plt.axes(projection='3d')
ax.set_box_aspect(aspect=None, zoom=0.9)
ax.set_title('Victims Rescued by Alpha and Beta for Stochastic Walk')
ax.set_xlabel('Alpha')
ax.set_ylabel('Beta')
ax.set_zlabel('Victims Rescued')
surf = ax.plot_trisurf(data['Alpha'], data['Beta'], data['Victims'] / VICTIMS, cmap=cm.coolwarm, edgecolor='none', linewidth=0, antialiased=False)
```

```python
data = data_weightedwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()
data['Victims'] -= data_randomwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()['Victims']
fig = plt.figure(figsize=(13.33,7.5), dpi = 96)
ax = plt.axes(projection='3d')
ax.set_box_aspect(aspect=None, zoom=0.9)
ax.set_title('delta Victims Rescued by Alpha and Beta for Greedy over Random')
ax.set_xlabel('Alpha')
ax.set_ylabel('Beta')
ax.set_zlabel('Victims Rescued')
surf = ax.plot_trisurf(data['Alpha'], data['Beta'], data['Victims'] / VICTIMS, cmap=cm.coolwarm, edgecolor='none', linewidth=0, antialiased=False)
```

```python
data = data_smgwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()
data['Victims'] -= data_randomwalk_64_32[['Alpha', 'Beta', 'Victims']].groupby(['Alpha', 'Beta']).mean().reset_index()['Victims']
fig = plt.figure(figsize=(13.33,7.5), dpi = 96)
ax = plt.axes(projection='3d')
ax.set_box_aspect(aspect=None, zoom=0.9)
ax.set_title('delta Victims Rescued by Alpha and Beta for Stochastic over Random')
ax.set_xlabel('Alpha')
ax.set_ylabel('Beta')
ax.set_zlabel('Victims Rescued')
surf = ax.plot_trisurf(data['Alpha'], data['Beta'], data['Victims'] / VICTIMS, cmap=cm.coolwarm, edgecolor='none', linewidth=0, antialiased=False)
```
